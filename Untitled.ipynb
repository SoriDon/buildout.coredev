```python
import pandas as pd
from pathlib import Path

pd.options.display.max_rows = 2000
data = pd.read_csv(Path("tests.csv"), names=["id", "layer", "package", "type"])
```

# Total number of tests

```python
len(data)
```

# Tests by type

```python
pd.pivot_table(data, index=["type"], values=["id"], aggfunc='count').sort_values('id', ascending=False)
```

# Tests by package

```python
pd.pivot_table(data, index=["package"], columns=["type"], values=["id"], aggfunc="count", fill_value=0, margins=True).sort_values(by=('id', 'All'), ascending=False)
```
